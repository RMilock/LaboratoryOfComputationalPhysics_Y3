## Fancy indexing

NumPy arrays can be indexed with slices, but also with boolean or integer arrays (masks). This method is called fancy indexing. It creates copies not views.

```python
a = np.random.randint(0, 21, 15) #15 rnd integeers from 0 to 21
print ("original array:", a,'\n')

mask = (a % 3 == 0) #ndarray of = shape but boolean
print ("the mask:", mask, type(mask),'\n')

filtered_a = a[mask]
# equivalent to a[a%3==0] 
print ("the filtered array:", filtered_a)

filtered_a=np.arange(15)
print ("the mod filtered array:", filtered_a)
print ("If false *mask* is creating a copy of a:", np.all(a==filtered_a))

# Indexing with a mask can be very useful to assign a new value to a sub-array:
a[a % 3 == 0] = -1 #applying to all the el so is valid  for every dim
print ("the modified array:", a,'\n')
```

## Broadcasting

So far we have seen elementwise operations on arrays, which works only on arrays of the same size/shape (Verify that out!!).

Nevertheless, It’s also possible to do operations on arrays of different
sizes if NumPy can transform these arrays so that they all have
the same size: this conversion is called broadcasting.

```python
#visualize what bradcasting does
from IPython.display import Image
Image("numpy_broadcasting.png")
```

```python
def decoro(f): 
    def wrapper(x):
        print("************")
        f(x) #print(x**2)
        print("************")
    return wrapper

@decoro
def imj(x): print(x)
    

imj("1st Image")
a = np.tile(np.arange(0, 40, 10), (3,1)).T #tile (A, multiply (row_times, column_times)-enanching)
b = np.tile(np.arange(0, 3), (4,1))
print(a+b)

#describe the 2nd image
imj("2nd Image")
print(np.arange(0, 40, 10))
print ("original array:",'\n',a,'\n')
print(np.arange(3))
print ("broadcasted sum :",'\n',a+np.arange(3),'\n')

#describe the 3rd image
imj("3nd Image")
a= np.arange(0,40,10).reshape(4,1)
b= np.arange(3)
print(a,b, a+b)
```
